```python
from bs4 import BeautifulSoup as soup
import requests
from splinter import Browser
from webdriver_manager.chrome import ChromeDriverManager
#in order for the browser to fully load
import time
import pandas as pd
import math
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.support import ui
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from selenium.common.exceptions import NoSuchElementException
```

```python
executable_path = {'executable_path': ChromeDriverManager().install()}
browser = Browser('chrome', **executable_path, headless=False)
```

```python
basic_url = 'https://www.imdb.com/search/title/?'
title_type = "feature,tv_movie"
release_date="2020-01-01,2020-12-31"
countries="us"
languages="en"
```

```python
target_url = f"{basic_url}title_type={title_type}&release_date={release_date}&countries={countries}&languages={languages}"
# print(target_url)
```

```python
browser.visit(target_url)
```

```python
html = browser.html
movie_soup = soup(html, 'html.parser')
range_titles = movie_soup.find_all('div', class_="desc")
result = [span.text.strip() for span in range_titles]
test = result[0].split()
number_of_titles_text = test[2]
number_of_titles = number_of_titles_text.replace(",","")

print(number_of_titles)
```

```python
count = 250
page = int(number_of_titles)/count
# times = 1

times = math.ceil(page)
start = 0

title_id=[]
title=[]
movie_genre=[]
movie_rating=[]
n_reviewer=[]

for n in range(times):
    start = count*n+1
    url=f"{basic_url}title_type={title_type}&release_date={release_date}&countries={countries}&languages={languages}&count={count}&start={start}&ref_=adv_nxt"
#     print(url)
    browser.visit(url)
    html = browser.html
    movie_soup = soup(html, 'html.parser')
    range_m=movie_soup.find_all('div', class_="lister-item-content")
#     print(range_m)
    for j in range(len(range_m)):
        print(j)
        try:
            soup_id = movie_soup.find_all('div', class_="ribbonize")[j].get('data-tconst')
  
            soup_refined=movie_soup.find_all('div',class_="lister-item-content")[j]
            soup_title=soup_refined.find("a").string
          
            soup_rating=soup_refined.find('div',class_="inline-block ratings-imdb-rating").get("data-value")
  
            soup_n_reviewer=soup_refined.find_all('meta')[2].get("content")
 
            soup_genre=soup_refined.find('span',class_="genre").text
           
            title_id.append(soup_id)
            title.append(soup_title)
            movie_rating.append(soup_rating)
            n_reviewer.append(soup_n_reviewer)
            movie_genre.append(soup_genre)
            
        except AttributeError as e:
                print(e)
    
```

```python
df_movies=pd.DataFrame(list(zip(title_id,title,movie_rating, n_reviewer, movie_genre)))
```

```python
df_movies[4]=df_movies[4].map(lambda x: x.lstrip('\n'))
df_movies[4]=df_movies[4]
df_movies
```
